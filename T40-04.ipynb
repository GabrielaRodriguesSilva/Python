# Estrutura de dados
## Listas (_``list``_)

É uma estrutura de dados que possui nenhum, um ou mais elementos ou itens. Os itens são ordenados a partir de um índice, apresentando flexibilidade para manipulação dos dados como inclusão, exclusão, alteração, ordenação, etc.

As listas são estruturas dinâmicas, ou seja, permitem que as operações possam ser realizadas em tempo de execução. Outro detalhe importante é que os dados armazenados em uma lista podem ser de tipos variados, permitindo um maior controle sobre os mesmos.

Em Python uma lista é representada por um par de colchetes ``[ ]``

```python
# Relembrando
x = 10

#Como saber o tipo do dado armazenado na variável x?
print(type(x))
```

# Operações com Listas

```python
lista = []
```

```python
# Adicionar via .append(value) - sempre no final da lista
lista.append('Banana')
lista.append('Abacaxi')
lista
```

```python
# Adicionar via insert(index, value) - insere em uma posição específica
lista.insert(1, 'Maça')
lista
```

```python
# Remover um item específico
lista.remove('Banana')
lista
```

```python
lista.append('Caqui')
lista.append('Cebola')
lista.append('Espinafre')
lista
```

```python
# Remoção do último indice (item) - pop()
print(lista)
lista.pop()
print(lista)
```

```python
#Removendo com pop(index)
lista.pop(1)
lista
```

```python
# Atividade para esquentar
# Leia via teclado 10 entradas numéricas (inteiros) "não repetidos"
# Ao final mostrar os números lidos
# Tempo: 10 minutos

lista = []

while len(lista) < 11: 
  val = int(input('Digite o valor: '))
  if (val) not in lista: # se não estiver na lista
    lista.append(val) # inserimos o número na lista
    print(val, 'inserido na lista')
  else:
    print(val, 'já foi inserido na lista anteriormente')
print(lista)
```

# Fatiamento de listas ou _``slice``_

```python
# O último índice é o tamanho da lista menos 1 -> len()-1
lista = ['Gabriela', 'Ana', 'Luiz', 'Pedro', 'Carol', 'Tio Patinhas', 'Pica Pau']
len(lista)
```

```python
# lista[início, fim, passo (número de contagem)]
#lista[start, stop, step]

# Forma 1 
lista[0:2] 
```

```python
# Forma 1 
print(lista[0:2])
print(lista[2:5])
```

```python
# Forma 1 
print(lista[0:2])
print(lista[2:5])

print(lista[3:]) #começa do 3 e vai até o final
print(lista[:3])
```

```python
# Forma 1 
print(lista[0:2])
print(lista[2:5])

print(lista[3:]) #começa do 3 e vai até o final
print(lista[:3])

print(lista[0:6:2]) #pulando
```

```python
# Forma 1 
print(lista[0:2])
print(lista[2:5])

print(lista[3:]) #começa do 3 e vai até o final
print(lista[:3])

print(lista[::2]) #pulando
```

```python
#Forma 2
v1 = slice(3) #slice (Vai pegar o valor e cortar esse índice)
print(lista[v1])
```

# Atividade (Legalzinha)
Elaborar um programa, utizizando todos os recursos aprendidos até o momento, que receba um conjunto de 10 entradas, contendo: nome, idade, peso.

A saída deverá conter:
* Maior idade
* Menor idade
* Média das iadades
* Média dos pesos


```python
#Entrada

# Edson, 49, 109
# Maria, 35, 64
# Jorge, 19, 95
# Ana, 23, 48

```

```python
pessoas = [
           
    ['Edson', 49, 114, 1.76],
    ['Maria', 35, 64, 1.62],
    ['Jorge', 18, 95, 1.84],
    ['Ana', 23, 48, 1.52]
    
]

idades = []
pesos = []
for pessoa in pessoas:
  idades.append(pessoa[1])
  pesos.append(pessoa[2])

print('Maior......:', max(idades))
print('Menor......:', min(idades))
print('Média idade:', sum(idades)/len(idades))
print('Média pesos:', sum(pesos)/len(pessoas))
```

```python
# Saída

# Maior......: 
# Menor......: 
# Média idade: 
# Média peso.: 
```

```python
# Resolução 2
pessoas = [
    ['Edson', 49, 114, 1.76],
    ['Maria', 35, 64, 1.62],
    ['Jorge', 18, 95, 1.84],
    ['Ana', 23, 48, 1.52]
]

quantidade = len(pessoas)
valores = []

for pessoa in pessoas:
  valores.insert(0, pessoa[1])
  valores.append(pessoa[2])

print('Maior......:', max(valores[:quantidade]))
print('Menor......:', min(valores[:quantidade]))
print('Média idade:', sum(valores[:quantidade])/quantidade)
print('Média pesos:', sum(valores[quantidade:])/quantidade)
```
