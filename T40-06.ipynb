* ``axis``: ``0`` para linhas e ``1`` para colunas
* ``inplance``: TRue, altera o DataFrame

# Filtragem de dados
* ``axis``: ``0`` para linhas e ``1`` para colunas

* ``DataFrame['Coluna'] = retorna uma Series``
* ``DataFrame[['Coluna']] = retorna um DataFrame``
*``DataFrame[['Coluna_1, 'Coluna_2']] = multíplas Colunas``
* ``DataFrame.query()``

````python
# Selecionar uma coluna (Series)
df_covid['Country/Region']
````

````python
# Selecionar uma coluna (DataFrame)
df_covid[['Country/Region']]
````

````python
# Mostrando apenas os cinco primeiros do filtro
df_covid[['Country/Region']].head()
````

````python
# E com mais colunas?
df_covid[['Country/Region', 'Confirmed', 'Deaths']].head()
````

````python
# Só para esquentar....
df_result = df_covid[['Country/Region', 'Confirmed', 'Deaths']]
df_result.plot()
````

````python
# Só para esquentar....
df_result = df_covid[['Country/Region', 'Confirmed', 'Deaths']].sample(2000)
df_result.plot()
````

````python
df_result = df_covid[['Deaths']].sample(2000)
df_result.plot()
````

````python
df_result = df_covid[['Deaths']].sample(2000)
df_result.plot.box()
````

````python
# Filtrando com mais parâmentros
df_covid.query('Confirmed > 80000 and Deaths > 15000').plot.bar()
````
