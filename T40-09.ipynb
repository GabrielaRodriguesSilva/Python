```python
# Importação das bibliotecas
import pandas as pd # para dados
import numpy as np # trabalhar com arrays numéricos
import matplotlib.pyplot as plt # biblioteca gráfica padrão, usada pelo Pandas
import seaborn as sns # outra biblioteca gráfica

# mostra os gráficos, suprimindo mensagens
%matplotlib inline 
```

```python
# carregamento do Dataset
mtcars = pd.read_csv('http://emsapi.esy.es/datasets/mtcars.csv', index_col=0)
```

# Características
mpg: Milhas por Galão (consumo)

cyl: Número de cilíndros

disp: Cilindada (em polegada cúbica)

hp: Cavalos de Potência (HP)

drat: Relação do eixo traseiro

wt: Peso em (1,000 libras)

qsec: Tempo que atinge 400m (1/4 de milha)

vs: Motor (0 = Forma em V, 1 = Reto)

am: Transmissão (0 = Automático, 1 = Manual)

gear: Número de marchas

carb: Número de carburadores

# Começando com gráficos

```python
# configurando o tamanho da figura (em polegadas)
plt.figure(figsize=(20, 18))
```

```python
# Usando o Pandas, determinando qual coluna deve ser utilizada
plt.figure(figsize=(15, 10))
mtcars['mpg'].plot()

# mostra o gráfico no notebook
plt.show()
```

```python
plt.figure(figsize=(10, 8))
mtcars['cyl'].plot()

# mostra o gráfico no notebook
plt.show()
```

```python
def grafico(DataFrame, coluna):
  plt.figure(figsize=(10, 8))
  DataFrame[coluna].plot()
  plt.show()
```

```python
grafico(mtcars, 'gear')
```

```python
notas = pd.read_csv('http://emsapi.esy.es/datasets/notas_alunos.csv', index_col=0)
```

```python
# Salvando a figura
plt.figure(figsize=(10, 8))
mtcars['cyl'].plot()

# salvamento
plt.savefig('figura01.png', dpi=300)
# mostra o gráfico no notebook
plt.show()
```

```python
# Dados sobre a distribuição dos dados (Histograma)
mtcars['qsec'].plot(kind='hist')
plt.title('Tempo que percorre 400m') # coloca um título no gráfico
plt.show()
```

```python
# Dados sobre a distribuição dos dados (Histograma)
mtcars['qsec'].plot(kind='bar')
plt.title('Tempo que percorre 400m') # coloca um título no gráfico
plt.show()
```

```python
# Incrementando o gráfico do histograma com uma densidade
mtcars['qsec'].plot(kind='density', lw=3)
mtcars['qsec'].plot(kind='hist', density=True)

plt.title('Tempo que percorre 400m') # coloca um título no gráfico
plt.show()
```
