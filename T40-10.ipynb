```python
# Juntando as coisa em um lugar só e personalizando
x = np.linspace(-2, 2, 100)

plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadrática')
plt.plot(x, x**3, label='Cúbica')

plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Juntando as coisa em um lugar só e personalizando...
x = np.linspace(-2, 2, 100)

plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadrática')
plt.plot(x, x**3, label='Cúbica')

plt.xlabel('Rótulo do eixo X')
plt.ylabel('Rótulo do eixo Y')
plt.title('Exemplo de Gráfico com Personalização')
plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Exagerando nos lables
x = np.linspace(-2, 2, 100)

plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadrática')
plt.plot(x, x**3, label='Cúbica')

plt.xlabel('Rótulo do eixo X')
plt.ylabel('Rótulo do eixo Y, este texto é grande mesmo, \nsó para exemplificar, quero mais texto, pois estou \ncom vontade de escrever')
plt.title('Exemplo de Gráfico com Personalização')
plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Personalizando os lables
x = np.linspace(-2, 2, 100)

plt.plot(x, x, label='Linear', color='red')
plt.plot(x, x**2, label='Quadrática', color='blue')
plt.plot(x, x**3, label='Cúbica', color='green')

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y')

plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Limites do gráfico
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, label='Linear', color='red')
plt.plot(x, x**2, label='Quadrática', color='blue')
plt.plot(x, x**3, label='Cúbica', color='green')

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y')

plt.xlim( (start, end) ) # limite do gráfico (limites de x e y)

plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Trabalhando com os ticks
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, label='Linear', color='red')
plt.plot(x, x**2, label='Quadrática', color='blue')
plt.plot(x, x**3, label='Cúbica', color='green')

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y')
plt.xlim( (start, end) ) # limite do gráfico (limites de x e y)
plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.xticks(ticks=np.linspace(start, end, 9),
           rotation=45) # espaço entre os valores dos rótulos (x)
plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Trabalhando com os ticks personalizados
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, label='Linear', color='red')
plt.plot(x, x**2, label='Quadrática', color='blue')
plt.plot(x, x**3, label='Cúbica', color='green')

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y')
plt.xlim( (start, end) ) # limite do gráfico (limites de x e y)
plt.legend(fontsize='x-large') # incluindo legenda no gráfico

plt.xticks(ticks=np.linspace(start, end, 9),
           rotation=45) # espaço entre os valores dos rótulos (x)

plt.yticks(
    ticks=[-8, -5, -2, 0, 2, 5, 8],
    labels=['BUM', 'ZOO', 'HOUSE', 'ZERO', 'BABA', 'CUCA', 'ZEBRA']
    )

plt.grid(True) # coloca um grid no gráfico
plt.show()
```

```python
# Mais customizações (personalizando as linhas)
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, label='Linear', color='red', linewidth=4)
plt.plot(x, x**2, label='Quadrática', color='blue')
plt.plot(x, x**3, label='Cúbica', color='green')

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y', style='italic', size=14)
plt.legend(fontsize=12)

plt.grid(True)
plt.show()
```

```python
# Mais customizações (personalizando as linhas, parte dois)
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, 'r--', label='Linear', linewidth=4, alpha=0.8)
plt.plot(x, x**2, 'b-.', label='Quadrática', linewidth=2)
plt.plot(x, x**3, 'g:', label='Cúbica', linewidth=3)

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y', style='italic', size=14)
plt.legend(fontsize=12)

plt.grid(True)
plt.show()
```

```python
from matplotlib.patches import ArrowStyle
# Fazendo anotações em gráficos
start = -2
end = 2

x = np.linspace(start, end, 100)

plt.plot(x, x, 'r--', label='Linear', linewidth=4, alpha=0.8)
plt.plot(x, x**2, 'b-.', label='Quadrática', linewidth=2)
plt.plot(x, x**3, 'g:', label='Cúbica', linewidth=3)

plt.title('Exemplo de Gráfico com Personalização', size=16)
plt.xlabel('Rótulo do eixo X', style='italic', size=14)
plt.ylabel('Rótulo do eixo Y', style='italic', size=14)
plt.legend(fontsize=12)

plt.annotate('Aqui', xy=(0.5, 4), color='red') # anotação em vermelho
plt.annotate('Acolá', 
             xy=(0.3, 4.5),
             xytext=(1, 6.0),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue'))

plt.grid(True)
plt.show()
```

# Múltiplos Plots

[Documentação](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot)

`fig, axes = plt.subplot(M, N)` 

* M são linhas
* N são colunas



```python
fig, axes = plt.subplots(1, 3, sharey='row')
x = np.linspace(0, 2, 100)

# Definindo o primeiro eixo
axes[0].plot(x, x, 'r--')
axes[0].set_title('Linear')
axes[0].set_xlabel('Eixo X')
axes[0].set_ylabel('Eixo Y')

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, sharey='row')
x = np.linspace(0, 2, 100)

# Definindo o primeiro eixo
axes[0].plot(x, x, 'r--')
axes[0].set_title('Linear')
axes[0].set_xlabel('Eixo X')
axes[0].set_ylabel('Eixo Y')

# Monte os gráficos para os demais índices (Quadrático e Cúbico)
axes[1].plot(x, x**2, 'b-.')
axes[1].set_title('Quadrática')
axes[1].set_xlabel('Eixo X')

axes[2].plot(x, x**3, 'g:')
axes[2].set_title('Cúbica')
axes[2].set_xlabel('Eixo X')

fig.suptitle('Título da Figura', fontsize=16) # suptitle porque a figura é matricial
plt.tight_layout() # Ajusta o padding dos subplots
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, sharey='col')
x = np.linspace(0, 2, 100)

# Definindo o primeiro eixo
axes[0].plot(x, x, 'r--')
axes[0].set_title('Linear')
axes[0].set_xlabel('Eixo X')
axes[0].set_ylabel('Eixo Y')

# Monte os gráficos para os demais índices (Quadrático e Cúbico)
axes[1].plot(x, x**2, 'b-.')
axes[1].set_title('Quadrática')
axes[1].set_xlabel('Eixo X')

axes[2].plot(x, x**3, 'g:')
axes[2].set_title('Cúbica')
axes[2].set_xlabel('Eixo X')

fig.suptitle('Título da Figura', fontsize=16) # suptitle porque a figura é matricial
plt.tight_layout() # Ajusta o padding dos subplots
plt.show()
```
