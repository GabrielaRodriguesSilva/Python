Faça um programa que leia o peso e a altura de uma pessoa, calcule o seu IMC e mostre o seu estado nutricional.

Observações:

Utilizar if-elif-else
O IMC deverá conter apenas duas casas decimais

Equação para o cálculo

IMC=pesoaltura2 

Tabela do Estado Nutricional

IMC	Resultado

Menos do que 18,5	Abaixo do peso

Entre 18,5 e 24,9	Peso normal

Entre 25 e 29,9	Sobrepeso

Entre 30 e 34,9	Obesidade grau 1

Entre 35 e 39,9	Obesidade grau 2

Mais do que 40	Obesidade grau 3


```python
peso = float(input('Digite seu peso: '))
altura = float(input('Digite sua altura: '))
imc = peso / (altura * altura)
print('IMC:', imc)

status = ''
if imc <= 18.5:
    status = 'Abaixo do peso'
elif (imc > 18.5 and imc < 24.9):
    status = 'Peso normal'
elif (imc > 25 and imc < 29.9):
    status = 'Sobrepeso'
elif (imc > 30 and imc < 34.9):
    status = 'Obesidade grau 1'
elif (imc > 35 and imc < 39.9):
    status = 'Obesidade grau 2'
else:
    status = 'Obesidade grau 3'

# função de arredondamento - round(valor, decimal)
print('IMC:', round(imc, 2), '(', status, ')')
```

# Estrutua de Dados: ''list''

Lista ou ''list'' é uma estrutura de dados que possui: _nenhum_, _um_ ou _mais_ itens que são ordenados internamente através de um índice. Este índice _sempre_ inicia em zero (0).

É uma estrutura flexível, ou seja, podemos adicionar, remover e alterar os valores contidos nela em tempo de execução.

Em Pyton uma lista é representada por uma sequência de itens separados por vírgula e dentro de _[]_ colchetes.


Os tipos de dados aceitos são dinâmicos. Assim, podem aceitar múltiplos tipos dentro de uma mesma lista.

```python
#Exemplos
lista = []
lista
```

```python
nomes = ['Gabriela', 'Giovana', 'Francisco', 'Enrique', 'Jonas', 'Yuri']
nomes
```

```python
lista_mista = ['Uninove', 3, 34.2, None, True]
lista_mista
```

```python
# Acessando elementos de uma lista
# lista [<índice] - em python índice se inicia em 0
nomes = ['Gabriela', 'Giovana', 'Francisco', 'Enrique', 'Jonas', 'Yuri']
print(nomes[0]) #primeiro elemento
print(nomes[3]) #terceiro elemento
```

```python
#Como mostrar todo o conteúdo da lista um a um?
for nome in nomes:
  print(nome)
```

```python
# Percorrendo a lista e mostrando a posição do elemento
for nome in nomes:
  print('O índice:', nomes.index(nome), 'é', nome) 
```

```python
# Pertecimento: operadores que calculam sobre os "valores" e não índices
#         0   1   2   3   4
numeros = [1, 2, 3, 4, 5]
print(3 in numeros)
print(6 in numeros)
print(7 not in numeros)
```
